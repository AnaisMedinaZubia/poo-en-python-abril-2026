#### Conceptos de OOPs (Object-Oriented Programming) Concepts in Python

lenguaje python, crear la clase ¨animal¨ con los atributos "nombre", "raza" crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python

# 🏗️ Definición de la Clase Animal
class Animal:
    # 🐣 El Constructor: ¡Donde la magia comienza!
    def __init__(self, nombre, raza):
        self.nombre = nombre  # 🏷️ Nombre asignado
        self.raza = raza      # 📝 Tipo de raza

    # 🗣️ Método Ladrar: ¡Aumenta el volumen!
    def ladrar(self):
        print(f"🐕 {self.nombre} ({self.raza}) lanza un: ¡GUAU! 📢")

    # 🏎️ Método Correr: ¡Activa los motores!
    def correr(self):
        print(f"⚡ {self.nombre} está corriendo... 💨")

# --- 🧪 Creación de los 3 Objetos (Instancias) ---

# 🦴 1. El Pitbull imponente
pitbull = Animal("Rambo 🥊", "Pitbull")

# 🌶️ 2. El Chihuahua valiente
chihuahua = Animal("Nacho 🌮", "Chihuahua")

# 🏘️ 3. El Perro Callejero (el alma del barrio)
callejero = Animal("Solovino 🦴", "Callejero")

# --- 🎬 ¡Acción en el parque! ---

print("🌳 --- ENCUENTRO EN EL PARQUE --- 🌳\n")

# 🐕 Acciones del Pitbull
pitbull.ladrar()
pitbull.correr()

print("-" * 25)

# 🐕 Acciones del Chihuahua
chihuahua.ladrar()
chihuahua.correr()

print("-" * 25)

# 🐕 Acciones del Perro Callejero
callejero.ladrar()
callejero.correr()

print("\n🏆 ¡Todos los perros han terminado su rutina! ✨")
print("\n✨ ¡Fin del entrenamiento! ✨")
```
