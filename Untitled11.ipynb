#### Conceptos de OOPs (Object-Oriented Programming) Concepts in Python

lenguaje python, crear la clase ¨animal¨ con los atributos "nombre", "raza" crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# 🏗️ Definición de la Clase Maestra
class Animal:
    # 🐣 El Constructor: Aquí es donde nace el animal
    def __init__(self, nombre, raza):
        self.nombre = nombre  # ✨ Nombre del peludo
        self.raza = raza      # 🧬 ADN del animal

    # 🗣️ Método para hacer ruido
    def ladrar(self):
        print(f"🐶 {self.nombre} el {self.raza} dice: ¡GUAU GUAU! 🔊")

    # 🏎️ Método para la acción
    def correr(self):
        print(f"🐕💨 {self.nombre} ha salido disparado... ¡Míralo correr!")

# --- 🧪 Creación de Instancias (Objetos) ---

# 🦴 Creando al poderoso Pitbull
pitbull = Animal("Tyson 🥊", "Pitbull")

# 🌶️ Creando al pequeño Chihuahua
chihuahua = Animal("Frijolito 🌯", "Chihuahua")

# --- 🎬 ¡Que comience la acción! ---

print("🌟 BIENVENIDOS AL PARQUE CANINO 🌟\n")

# 🐕 Acciones del primer objeto
pitbull.ladrar()
pitbull.correr()

print("\n" + "🐾" * 15 + "\n")

# 🐕 Acciones del segundo objeto
chihuahua.ladrar()
chihuahua.correr()

print("\n✨ ¡Fin del entrenamiento! ✨")
```
