#### Conceptos de OOPs (Object-Oriented Programming) Concepts in Python

lenguaje python, crear la clase ¨animal¨ con los atributos "nombre", "raza" crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor: se ejecuta al crear el objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre  # Atributo nombre
        self.raza = raza      # Atributo raza

    # Función para ladrar
    def ladrar(self):
        print(f"{self.nombre} ({self.raza}) está ladrando: ¡Guau guau!")

    # Función para correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente.")
    def Morder(self):
        print(f"{self.nombre} está mordiendo al vecino.")
# --- Creación de los objetos (Instanciación) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = Animal("Zeus", "Pitbull")

# Creamos el objeto 'chihuahua' con datos específicos
chihuahua = Animal("Chispa", "Chihuahua")
# Creamos el objeto perro callejero con datos
perrocallejero = Animal("solovino", "Chafa")
# --- Uso de los objetos ---

# Acciones del Pitbull
pitbull.ladrar()
pitbull.correr()
pitbull.Morder()

print("-" * 20) # Separador visual

# Acciones del Chihuahua
chihuahua.ladrar()
chihuahua.correr()
chihuahua.Morder()

print("-" * 20) # Separador visual

# Acciones del perro callejero
print(F"Datos del tercer perro: {perrocallejero.nombre} ({perrocallejero.raza})")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.Morder()
print(" Programa realizado por Anais Medina NL = 36")
```
